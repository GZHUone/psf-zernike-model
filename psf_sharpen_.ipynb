```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image
image = cv2.imread('single_psf_mean_denoised_resized.png', cv2.IMREAD_GRAYSCALE)

# Ensure the image is loaded correctly
if image is None:
    print("Image loading failed!")
    exit()

# High-pass filter
def high_pass_filter(image):
    blurred = cv2.GaussianBlur(image, (5, 5), 0)  # Apply low-pass filter
    high_pass = image - blurred  # High-pass filter
    sharpened_image = image + high_pass  # Enhance image
    return sharpened_image

# Laplacian sharpening
def laplacian_sharpening(image):
    kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])  # Laplacian kernel
    sharpened_image = cv2.filter2D(image, -1, kernel)  # Apply the kernel
    return sharpened_image

# Unsharp mask
def unsharp_mask(image):
    blurred = cv2.GaussianBlur(image, (5, 5), 1.0)  # Apply Gaussian blur
    sharpened_image = cv2.addWeighted(image, 1.5, blurred, -0.5, 0)  # Apply unsharp mask
    return sharpened_image

# Apply the three sharpening methods
high_pass_sharpened = high_pass_filter(image)
laplacian_sharpened = laplacian_sharpening(image)
unsharp_sharpened = unsharp_mask(image)

# Define font properties for the title
font_properties = {'fontsize': 25, 'fontweight': 'bold', 'family': 'serif'}

# Display the results
fig, axes = plt.subplots(1, 3, figsize=(20, 5))

# High-pass filter sharpening result
axes[0].imshow(high_pass_sharpened, cmap='gray')
axes[0].set_title('High-pass Filter Sharpening', fontdict=font_properties)
axes[0].axis('off')

# Laplacian sharpening result
axes[1].imshow(laplacian_sharpened, cmap='gray')
axes[1].set_title('Laplacian Sharpening', fontdict=font_properties)
axes[1].axis('off')

# Unsharp mask result
axes[2].imshow(unsharp_sharpened, cmap='gray')
axes[2].set_title('Unsharp Mask', fontdict=font_properties)
axes[2].axis('off')

plt.tight_layout()
plt.show()
```
